### **Buscando filas duplicadas y eliminando filas duplicadas**

```python
filas_duplicadas = df.duplicated().sum()

if filas_duplicadas > 0:
    print(f'Se encontraron un total de {filas_duplicadas} filas duplicadas, eliminando filas duplicadas')
    df_sin_duplicados = df.drop_duplicates(inplace=True)
else:
    print('No se encontraron filas duplicadas')
```

### **Buscando y eliminando columnas y filas con Nulos**

**Columnas con Nulos**

```python
columnas_con_nulos = df.isnull().sum()
columnas_con_nulos
```

**Fila con Nulos**

```python
filas_con_nulos = df[df.isnull().any(axis=1)]
filas_con_nulos
```

**Total de datos con Nulos**

```python
print(f"Se encontraron un total de {columnas_con_nulos.sum()} datos nulos")
```

**Eliminando Datos Nulos**

```python
df_sin_nulos = df.dropna()
df_sin_nulos
```

### **Filtrando Pokemons de tipo Fuego y mostrando los primeros 20**

```python
def filtro_tipo_fuego(Type_1):
    return "Fire" in Type_1

solo_tipo_fuego = df[df['Type 1'].apply(filtro_tipo_fuego)]

solo_tipo_fuego.head(20)
```

```python
promedio_defensa = solo_tipo_fuego['Def'].mean()

print(f'El promedio de la defensa de todos los pokémon de tipo Fuego es de {promedio_defensa} puntos de defensa')
```

### **Creando Gráficos**

**Gráfico de barras**

**Top 10 de pokemons tipo fuego**

```python
nombre_ataque_fuego = solo_tipo_fuego[['Name', 'Att']]
```

```python
top_10_fuego = nombre_ataque_fuego.head(10)

top_10_fuego.set_index("Name", inplace=True)

top_10_fuego["Att"].plot(kind='bar')
plt.title("Top 10 Pokémon de tipo Fuego (en base a su ataque) ")
plt.ylabel("Puntos de Ataque")
plt.xlabel("Nombre de Pokémons")

plt.show()
```

**Gráfico de Líneas**

```python
generacion_pokemon_nombre = solo_tipo_fuego['Name']
generacion_pokemon = solo_tipo_fuego['Generation']
```

```python
plt.plot(generacion_pokemon, marker ='o')
plt.plot(generacion_pokemon_nombre, marker ='o')

plt.show()
```
